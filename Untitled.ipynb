```python
import pandas as pd
import pulp
from itertools import product
import numpy as np
```

```python
dict_base = df2.to_dict()
dict_base
```

```python
# GENERAZIONE DIZIONARI PARAMETRI
#names = dict_base['Denominazione']
codes = dict_base['Codice']  # solo per output
semester = dict_base['Sem']
cfus = dict_base['CFU']
group_full = dict_base['Gruppo']
#group_dic = dict_base['Gruppo_id']
rating = dict_base['Rating']
min_cfus = [10, 16, 10, 16, 0, 0] # TODO È FISSO :(

group_full
```

```python
# INSIEMI
# TODO far sì che queste cose siano dinamiche in base al source
COURSES = list(codes.keys())
YEARS = [1, 2]


GROUPS_names = df['Gruppo'].unique()
GROUPS = np.arange(len(GROUPS_names))
GROUPS
```

```python
# PARAMETRI
# combos = [(i, j) for i in COURSES for j in YEARS]
# combos = product(COURSES, YEARS, GROUPS)

# PROBLEMA
prob = pulp.LpProblem('Piano_di_studi', pulp.const.LpMaximize)
x = pulp.LpVariable.dicts('corso-anno-gruppo', (COURSES, YEARS, GROUPS), lowBound=0, upBound=1, cat='Continuous')  # creazione variabili
y = pulp.LpVariable.dicts('corso',             (COURSES),                lowBound=0, upBound=1, cat='Binary')  # 1 se corso i è scelto
#z = pulp.LpVariable.dicts('corso-anno',        (COURSES, YEARS),         lowBound=0, upBound=1, cat='Binary')  # 1 se corso i all'anno j
```

```python
# FUNZIONE OBIETTIVO
prob += pulp.lpSum([rating[i]*cfus[i]*x[i][j][k] for (i, j, k) in product(COURSES, YEARS, GROUPS)])
```

```python
CFU_max_sem=30
CFU_max_tot=60 # il max di quelli a scelta (obbligatori sono 68)
```

```python
# VINCOLI

# vincolo logico di bigM

# se un corso è scelto allora z_ij = 1
#for (i,j) in product(COURSES,YEARS):
#    prob += pulp.lpSum([x[i][j][k] for k in GROUPS]) <= 4000 * z[i][j]

# non posso scegliere un corso in più anni (cioè mettere GAT metà al primo e metà al secondo annno per es.)
#for i in COURSES:
#    prob += pulp.lpSum([z[i][j] for j in YEARS]) <= 1

# se un corso è scelto allora y[i]=1
for i in COURSES:
    prob += pulp.lpSum([x[i][j][k] for (j, k) in product(YEARS, GROUPS)]) <= 1000 * y[i]

# y[i]=1 allora tutti quanti sopra tutti i gruppi devono sommare a 1 (circa)
for i in COURSES:
    prob += y[i] <= pulp.lpSum([x[i][j][k] for (j, k) in product(YEARS, GROUPS)]) - 0.001

# corso scelto al massimo una sola volta
#for i in COURSES:
#    prob += pulp.lpSum([x[i][j][k] for (j, k) in product(YEARS, GROUPS)]) <= 1, (f"Corso {i} scelto una sola volta")
```

```python
# max cfu per semestre per anno
#for j in YEARS:
#    group_courses=[i for i in COURSES if semester[i] == 1]
#    prob += pulp.lpSum(
#        [cfus[i]*x[i][j][k] for (i,k) in product(COURSES,GROUPS)]) <= CFU_max_sem, ("Max cfu sem 1 anno " + str(j))
#
#    group_courses=[i for i in COURSES if semester[i] == 2]
#    prob += pulp.lpSum(
#        [cfus[i]*x[i][j][k] for (i,k) in product(COURSES,GROUPS)]) <= CFU_max_sem, ("Max cfu sem 2 anno " + str(j))

# min cfu gruppo
for k in GROUPS:
    prob += pulp.lpSum([cfus[i]*x[i][j][k] for (i,j) in product(COURSES,YEARS)]) == min_cfus[k], (f"Min cfu gruppo {k}")  # >=

# max cfu totali
prob += pulp.lpSum([cfus[i]*y[i] for i in COURSES]) <= CFU_max_tot, "Max cfu totali"
```

```python
# dfcrosstab.loc[52503]['FREE']
# dfcrosstab.loc[codes[...]][GROUPS_names[...]]


# posso dare peso a un corso in un gruppo solo è consentito
for (i,k) in product(COURSES,GROUPS):
    prob += pulp.lpSum([x[i][j][k] for j in YEARS]) <= dfcrosstab.loc[codes[i]][GROUPS_names[k]]
```

```python
# RISOLUZIONE
status = prob.solve()
```
